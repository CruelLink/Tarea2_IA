```python
dffinal = pd.DataFrame()
for y in df.columns[ 2 :  ]:
    dftemp  = df.loc[ :  , ['Country Name','Country Code', y] ]
    dftemp['Year'] = y
    dftemp.columns = ['Country Name', 'Country Code', 'Value', 'Year']
    dftemp = dftemp.loc[ : , ['Country Name', 'Country Code', 'Year', 'Value'] ]
    
    dffinal = pd.concat([dffinal, dftemp])
```

```python
df_countries_data= pd.merge(df_population,df_continents, left_on='Country (or dependency)', right_on='name', how = 'left')
```

```python
df_countries_data.to_csv('df_countries_data.csv', index = False)
```

```python
df_Final_data= pd.merge(df,df_countries_data, left_on='Country Name', right_on='Country (or dependency)', how = 'left')
```

```python
df_Final_data['Population (2020)']
```

```python
dffinal = pd.DataFrame()
for y in df.columns[ 2 :  ]:
    dftemp  = df_Final_data.loc[ :  , ['Country Name','Country Code', y ,'region', 'Population (2020)'] ]
    dftemp['Year'] = y
    dftemp.columns = ['Country Name', 'Country Code', 'Value','region','Population (2020)','Year']
    dftemp = dftemp.loc[ : , ['Country Name', 'Country Code','Year', 'Value', 'region', 'Population (2020)'] ]
    
    dffinal = pd.concat([dffinal, dftemp])
```
